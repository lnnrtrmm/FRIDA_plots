### This example uses the carbon balance data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time_FRIDA = np.linspace(1980,2130,151)


filename = 'Carbon_Balance/FRIDA_historical_budget.csv'
df = pd.read_csv(filename, delimiter=';')

atm_co2_growth_FRIDA = df.values[:,1].astype(float)
Emis_flu_FRIDA    = df.values[:,2].astype(float)
Emis_other_FRIDA  = df.values[:,3].astype(float)
Emis_energy_FRIDA = df.values[:,4].astype(float)
land_sink_FRIDA   = df.values[:,5].astype(float)
ocean_sink_FRIDA  = df.values[:,6].astype(float)
```

```python
font=14
plt.figure(figsize=(12,8), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)

ax=plt.gca()

ax.stackplot(time_FRIDA, Emis_flu_FRIDA, Emis_energy_FRIDA+Emis_other_FRIDA, 
             labels=['Land use change', 'Fossil fuels and industry'], colors=['gold', 'grey'],
             alpha=1.0, edgecolor='k')
ax.stackplot(time_FRIDA, -ocean_sink_FRIDA, -land_sink_FRIDA, -atm_co2_growth_FRIDA,
             labels=['Ocean', 'Land', 'Atmosphere'], colors=['lightseagreen', 'yellowgreen', 'skyblue'],
             edgecolor='k', alpha=1.0)


# Adding labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('CO$_2$-Flux (GtC yr$^{-1}$)')
ax.set_title('Annual carbon balance in FRIDA')
ax.set_xlim(1980,2100)
ax.set_ylim(-15,20)

ax.grid()
ax.legend(ncol=2, loc='upper left')

# Show plot
plt.subplots_adjust(wspace=0.1)
plt.savefig('EXAMPLE_stacked_area_chart.png')
```
