```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from lib_FRIDA_data import *
```

```python
expID='EMB_13March2025_nS100000'
path='/home/lennart/Arbeit/2_WorldTrans/work/FRIDA_plots/'
atm_co2_growth_FRIDA, time_FRIDA = readFromFridaOutput(path, 'CO2growth', expID=expID, lmedian=True)
Emis_flu_FRIDA, time_FRIDA = readFromFridaOutput(path, 'CO2emis_FLU', expID=expID, lmedian=True)
Emis_other_FRIDA, time_FRIDA = readFromFridaOutput(path, 'CO2emis_other', expID=expID, lmedian=True)
land_sink_FRIDA, time_FRIDA = readFromFridaOutput(path, 'LandCO2uptake', expID=expID, lmedian=True)
Emis_energy_FRIDA, time_FRIDA = readFromFridaOutput(path, 'CO2emis_energy', expID=expID, lmedian=True)
ocean_sink_FRIDA, time_FRIDA = readFromFridaOutput(path, 'OceanCO2uptake', expID=expID, lmedian=True)


time_GCB = np.linspace(1750,2022,273)
df = pd.read_csv('GCB_historical_budget.csv', delimiter=',')
Emis_all_GCB = df.values[:,1]
Emis_flu_GCB = df.values[:,2]
atm_co2_growth_GCB = df.values[:,3]
ocean_sink_GCB = df.values[:,4]
land_sink_GCB = df.values[:,5]
cement_sink_GCB = df.values[:,6]
```

```python
font=14
plt.figure(figsize=(12,8), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)

# GCB data
ax=plt.gca()



ax.stackplot(time_FRIDA, Emis_flu_FRIDA, Emis_energy_FRIDA, Emis_other_FRIDA, labels=['Land use change', 'Fossil fuels', 'Concrete'], colors=['gold', 'grey', 'lightgrey'], alpha=1.0, edgecolor='k')
ax.stackplot(time_FRIDA, -ocean_sink_FRIDA, -land_sink_FRIDA, -atm_co2_growth_FRIDA, labels=['Ocean', 'Land', 'Atmosphere'], colors=['lightseagreen', 'yellowgreen', 'skyblue'], edgecolor='k', alpha=1.0)


ax.plot(time_GCB, Emis_flu_GCB, color='darkorange', linewidth=3)
ax.plot(time_GCB, Emis_flu_GCB+Emis_all_GCB-cement_sink_GCB, color='k', linewidth=3, label='GCB (Friedlingstein, 2023)')
ax.plot(time_GCB, -ocean_sink_GCB, color='teal', linewidth=3)
ax.plot(time_GCB, -ocean_sink_GCB-land_sink_GCB, color='green', linewidth=3)
ax.plot(time_GCB, -ocean_sink_GCB-land_sink_GCB-atm_co2_growth_GCB, color='royalblue', linewidth=3)


# Adding labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('CO$_2$-Flux (GtC yr$^{-1}$)')
ax.set_title(r'Annual carbon balance in FRIDAv2.1 beta 1')
ax.set_xlim(1980,2150)
ax.set_ylim(-25,30)

ax.grid()
ax.legend(ncol=2, loc='upper left')

# Show plot
plt.subplots_adjust(wspace=0.1)
plt.savefig('FRIDA_EMBv2.1Beta1_C_balance.png')
```
