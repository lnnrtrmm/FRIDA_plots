```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('FRIDA_historical_budget.csv', delimiter=';')


time_FRIDA = np.linspace(1980,2130,151)
atm_co2_growth_FRIDA = df.values[:,1].astype(float)
Emis_flu_FRIDA    = df.values[:,2].astype(float)
Emis_other_FRIDA  = df.values[:,3].astype(float)
Emis_energy_FRIDA = df.values[:,4].astype(float)
land_sink_FRIDA   = df.values[:,5].astype(float)
ocean_sink_FRIDA  = df.values[:,6].astype(float)


df = pd.read_csv('GCB_historical_budget.csv', delimiter=',')

time_GCB = np.linspace(1750,2022,273)
Emis_all_GCB = df.values[:,1]
Emis_flu_GCB = df.values[:,2]
atm_co2_growth_GCB = df.values[:,3]
ocean_sink_GCB = df.values[:,4]
land_sink_GCB = df.values[:,5]
cement_sink_GCB = df.values[:,6]
```

```python
font=12
plt.figure(figsize=(14,5), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)

# GCB data
plt.subplot(121)
ax=plt.gca()


ax.stackplot(time_GCB, Emis_flu_GCB, Emis_all_GCB-cement_sink_GCB, labels=['CO$_2$ emissions from Land-Use', 'Fossil CO$_2$ emissions'], colors=['gold', 'grey'])
ax.stackplot(time_GCB, -ocean_sink_GCB, -land_sink_GCB, -atm_co2_growth_GCB, labels=['Ocean sink', 'land sink', 'Atmospheric CO$_2$ growth'], colors=['darkcyan', 'yellowgreen', 'skyblue'])

# Adding labels and legend
ax.set_xlabel('Year')
ax.set_ylabel('CO$_2$-Flux (GtC yr$^{-1}$)')
ax.set_title(r'Annual carbon emissions and partitioning in $\bf{GCB}$')
ax.set_xlim(1850,2022)
ax.set_ylim(-15,15)

ax.grid()
ax.legend(loc='upper left')


plt.subplot(122)
ax=plt.gca()

ax.stackplot(time_FRIDA, Emis_flu_FRIDA, Emis_energy_FRIDA+Emis_other_FRIDA, labels=['CO$_2$ emissions from Land-Use', 'Fossil CO$_2$ emissions'], colors=['gold', 'grey'])
ax.stackplot(time_FRIDA, -ocean_sink_FRIDA, -land_sink_FRIDA, -atm_co2_growth_FRIDA, labels=['Ocean sink', 'land sink', 'Atmospheric CO$_2$ growth'], colors=['darkcyan', 'yellowgreen', 'skyblue'])

# Adding labels and legend
ax.set_xlabel('Year')
#ax.set_ylabel('CO$_2$-Flux (GtC yr$^{-1}$)')
ax.set_title(r'Annual carbon emissions and partitioning in $\bf{FRIDA}$')
ax.set_xlim(1980,2130)
ax.set_ylim(-15,15)

ax.grid()
#ax.legend(loc='upper right')

# Show plot
plt.subplots_adjust(wspace=0.1)
plt.show()
```
