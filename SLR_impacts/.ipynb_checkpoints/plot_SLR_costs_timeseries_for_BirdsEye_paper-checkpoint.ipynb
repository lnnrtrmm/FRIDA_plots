```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time_FRIDA = np.linspace(1980,2130,151)


filenames = ['do_nothing_costs.csv', 'protect_always_costs.csv', 'retreat_always_costs.csv', 
             'protect_later_costs.csv', 'retreat_later_costs.csv']

scenario_names = ['Do nothing scenario', 'Protect always', 'Retreat always', 'Protect after 2025', 'Retreat after 2025']
scenario_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:blue', 'tab:green']
scenario_styles = ['-', '-', '-', '--', '--']

# Data are annual costs and annual deaths, but accumulated land lost
SLR_money_costs_ann = []
SLR_lives_lost_ann = []
SLR_land_lost = []

for filename in filenames:
    df = pd.read_csv(filename, delimiter=';')

    SLR_lives_lost_ann.append(df.values[:,1].astype(float))
    SLR_land_lost.append(df.values[:,2].astype(float))
    SLR_money_costs_ann.append(df.values[:,3].astype(float))
```

```python
font=13
plt.figure(figsize=(8,8), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)

ax=plt.gca()

for i, scenario in enumerate(scenario_names):
    ax.plot(time_FRIDA, np.cumsum(SLR_money_costs_ann[i]), color=scenario_colors[i], label=scenario, linewidth=3,
           linestyle=scenario_styles[i])

ax.set_xlabel('Year')
ax.set_ylabel('cum. costs in Giga\$(2021)')
ax.set_title('Total costs of SLR')
ax.set_xlim(1980,2100)
ax.set_yscale('log')
ax.set_ylim(10,10**5)

ax.grid()
ax.legend(loc='lower right')


#plt.subplots_adjust(wspace=0.2)
plt.tight_layout()
#plt.show()
plt.savefig('FRIDA_SLR_costs_for_BirdsView_Paper.png')
```

```python
font=13
plt.figure(figsize=(8,8), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)


ax=plt.gca()

for i, scenario in enumerate(scenario_names):
    ax.plot(time_FRIDA, np.cumsum(SLR_lives_lost_ann[i])*1e6, color=scenario_colors[i], label=scenario, linewidth=3,
           linestyle=scenario_styles[i])

ax.set_xlabel('Year')
ax.set_ylabel('cum. fatalities (People)')
ax.set_title('Total flood fatalities from SLR')
ax.set_xlim(1980,2100)
#ax.set_ylim(-15,20)
ax.set_yscale('log')


ax.grid()
ax.legend()


#plt.subplots_adjust(wspace=0.2)
plt.tight_layout()
#plt.show()
plt.savefig('FRIDA_SLR_fatalities_for_BirdsView_Paper.png')
```

```python
font=13
plt.figure(figsize=(8,8), facecolor='white')
plt.rcParams.update({'font.size': font})
plt.rc('xtick', labelsize=font)
plt.rc('ytick', labelsize=font)


ax=plt.gca()

for i, scenario in enumerate(scenario_names):
    ax.plot(time_FRIDA, SLR_land_lost[i], color=scenario_colors[i], label=scenario, linewidth=3,
           linestyle=scenario_styles[i])


ax.set_xlabel('Year')
ax.set_ylabel('loss ( km$^2$)')
ax.set_title('Total loss of land')
ax.set_xlim(1980,2100)
#ax.set_ylim(-15,20)
ax.set_yscale('log')

ax.grid()
ax.legend()

#plt.subplots_adjust(wspace=0.2)
plt.tight_layout()
#plt.show()
plt.savefig('FRIDA_SLR_land_loss_for_BirdsView_Paper.png')
```
